## 3.1 Problem Definition
### *Objective of the learning task*
### The objective of this phase is to construct data-driven surrogate models for the pressure-strain correlation tensor appearing in the Reynolds Stress Transport Equations (RSTE), using supervised machine learning.
### In the Reynolds stress modeling framework, the pressure-strain correlation governs the redistribution of turbulent kinetic energy among Reynolds stress components, and is the most challenging term to model accurately using classical closure assumptions. This motivates its direct learning from high-fidelity DNS data.

### Target variables: pressure–strain correlation components

### For incompressible turbulent channel flow, the pressure–strain correlation tensor $\boldsymbol{\Phi}_{ij}$ is:

#### **Symmetric**: $\Phi_{ij} = \Phi_{ji}$
#### **Traceless**: $\Phi_{ii} = 0$

### Due to flow homogeneity in the streamwise and spanwise directions and statistical symmetries of plane channel flow, only **four independent components** are non-zero and physically relevant:

### $$[\Phi_{11}, \quad \Phi_{22}, \quad \Phi_{33}, \quad \Phi_{12}]$$

### Each of these components is modeled **independently** as a scalar regression problem:

### $$[\hat{\Phi}*{ij} = f*{ij}(\mathbf{x}), \quad ij \in {11,22,33,12}]$$

### This design choice is intentional:

#### * It avoids introducing artificial coupling constraints between components
#### * It mirrors the modeling strategy adopted in the reference study
#### * It simplifies training, optimization, and interpretability
#### * It allows component-wise error analysis in later validation

----

### Input features: physics-based modeling basis

### The input feature set is constructed from **classical turbulence modeling theory**, ensuring that the machine learning model respects known physical dependencies of the pressure-strain term.

### The selected inputs are:

### 1. Reynolds stress anisotropy tensor

### $$[b_{ij} = \frac{R_{ij}}{2k} - \frac{1}{3}\delta_{ij}]$$

### Components used:

### $$[b_{11}, \quad b_{22}, \quad b_{33}, \quad b_{12}]$$

### **Physical motivation**:

#### * Pressure–strain acts to redistribute anisotropy
#### * Classical closures (e.g. Rotta, SSG) explicitly depend on $b_{ij}$
#### * Anisotropy encodes directional turbulence structure not captured by scalar quantities

----

### 2. Turbulence kinetic energy

### $$[k = \frac{1}{2}(R_{11} + R_{22} + R_{33})]$$

### **Physical motivation**:

#### * Sets the turbulence energy scale
#### * Appears multiplicatively in rapid pressure–strain models
#### * Necessary for correct magnitude scaling of $\Phi_{ij}$

----

### 3. Dissipation rate

### $$[\varepsilon = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}]$$

### **Physical motivation**:

#### * Represents small-scale energy removal
#### * Controls near-wall behavior of pressure–strain
#### * Essential for stability and realizability of learned models

----

### 4. Mean velocity gradient

### $$[\frac{\partial U}{\partial y}]$$

### **Physical motivation**:

#### * Channel flow has only one non-zero mean velocity gradient
#### * Governs production of turbulence
#### * Appears explicitly in rapid pressure–strain formulations
#### * Encodes mean–turbulence interaction

----

### Functional mapping learned by ML models

### The supervised learning problem is formulated as:

### $$[\Phi_{ij} = f\left(b_{12},k,\varepsilon,\frac{\partial U}{\partial y}\right)]$$

### where:

### * $f(\cdot)$ is a nonlinear function approximated using:

  #### * Artificial Neural Networks (MLP)
  #### * Random Forest regression
  #### * Gradient Boosted Decision Trees

----

### Modeling philosophy

### This formulation adheres to the following principles:

#### * **Physics-informed feature selection** (no black-box inputs)
#### * **Galilean invariance** (features depend on gradients and tensors)
#### * **Local closure assumption** consistent with RANS
#### * **Compatibility with Reynolds stress transport modeling**

### By learning the pressure–strain correlation directly from DNS statistics, this approach seeks to overcome the structural limitations of traditional closure models while remaining compatible with RANS-level CFD solvers.

## 3.2 Dataset Organization and Case-wise Training Strategy

### Available Reynolds numbers

### The study employs Direct Numerical Simulation (DNS) data for fully developed turbulent plane channel flow at four friction Reynolds numbers:

### $$[Re_\lambda \in {550,1000,2000,5200}]$$

### These datasets span more than **one order of magnitude in Reynolds number**, providing a stringent test for both physical modeling and machine learning generalization.

### Each dataset consists of wall-normal profiles of turbulence statistics expressed in inner-scaled coordinates ($y^+$), including Reynolds stresses, dissipation, pressure–strain correlation, and mean velocity gradients.

---

### Motivation for multi–Re dataset splitting

### In turbulence modeling, **generalization across Reynolds numbers** is significantly more challenging than interpolation within a single flow regime. A model that merely fits a single Reynolds number may:

#### * Learn dataset-specific scaling artifacts
#### * Overfit near-wall behavior unique to that Reynolds number
#### * Fail to extrapolate to unseen turbulent states

### Therefore, model performance must be assessed not only within known data but also across **unseen Reynolds numbers**, which mirrors practical RANS usage.

### To enforce this, a **case-wise training strategy** is adopted, consistent with the reference study (Table 1).

---

### Case-wise training configuration (Table 1)

### Four independent training cases are constructed.
### In each case:

#### * **Three Reynolds numbers** are used for training and validation
#### * **One Reynolds number is completely excluded** from Phase 3
#### * The excluded Reynolds number is reserved **only for testing in Phase 4**

### $$[\text{Training} \cap \text{Testing} = \varnothing]$$

### The four cases are defined as follows:

| Case | Training $Re_\lambda$ | Held-out $Re_\lambda$ (Phase 4) |
| ---: | ------------------ | ---------------------------- |
|    1 | 550, 1000, 2000    | 5200                         |
|    2 | 550, 1000, 5200    | 2000                         |
|    3 | 550, 2000, 5200    | 1000                         |
|    4 | 1000, 2000, 5200   | 550                          |

#### *This table corresponds to Table 1 provided in the reference paper and stored locally.*

---

### Phase-wise data usage policy

### To avoid information leakage and ensure a clean evaluation pipeline, the following strict separation is enforced:

### * **Phase 3 (Training & Validation)**

  #### * Uses **only the training Reynolds numbers** listed for each case
  #### * Hyperparameter tuning is performed exclusively within these datasets
  #### * Validation data is drawn from the same Reynolds numbers as training

### * **Phase 4 (Testing & Generalization)**

  #### * Uses **only the held-out Reynolds number**
  #### * No retraining, tuning, or normalization recalculation is allowed

### This ensures that Phase 4 genuinely evaluates **Reynolds-number generalization**, not interpolation.

---

### Data aggregation within each case

### For a given case, all wall-normal samples from the selected training Reynolds numbers are **pooled together** to form a single training dataset:

### $$[\mathcal{D}_{\text{train}}^{(c)}=\bigcup_{Re_\lambda \in \mathcal{R}*{\text{train}}^{(c)}} \mathcal{D}(Re*\lambda)]$$

### where:

#### * $\mathcal{R}_{\text{train}}^{(c)}$ denotes the training Reynolds numbers for case $c$
#### * $\mathcal{D}(Re_\lambda)$ represents the DNS samples at a given Reynolds number

### This pooling strategy allows the model to:

#### * Observe turbulence physics across multiple Reynolds regimes
#### * Learn Reynolds-invariant trends under inner scaling
#### * Avoid memorizing a single $Re_\lambda$ profile

---

### Rationale for four independent training cases

### Training four independent cases serves two critical purposes:

#### 1. **Robustness assessment**
   #### Each Reynolds number is tested as an unseen case exactly once.

#### 2. **Physics-driven generalization analysis**
   #### Differences in performance across cases reveal:

   #### * Which Reynolds numbers are harder to extrapolate to
   #### * Where inner scaling succeeds or breaks down
   #### * How model capacity affects cross-Re learning

### This structure transforms the ML problem from simple regression into a **physics-informed generalization study**.

---

## **3.3 Train-Validation Split Strategy**

### Purpose of the train-validation split

### The purpose of the train-validation split in Phase 3 is to:

#### * Tune model hyperparameters
#### * Monitor overfitting during training
#### * Select models that generalize **within known Reynolds-number regimes**

### Importantly, the validation set **must not** be used to assess Reynolds-number generalization.
### That objective is **strictly reserved for Phase 4**.

---

### Key principle: No Reynolds-number leakage

### To preserve the physical meaning of the learning task, **Reynolds-number information must not leak** between training, validation, and testing phases.

### Formally,

### $$[\mathcal{D}*{\text{train}} \cap \mathcal{D}*{\text{test}} = \emptyset]$$

### $$[\mathcal{D}*{\text{val}} \subset \mathcal{D}*{\text{train-Re}}]$$

### where:

#### * $( \mathcal{D}_{\text{train}} )$: training data
#### * $( \mathcal{D}_{\text{val}} )$: validation data
#### * $( \mathcal{D}_{\text{test}} )$: held-out Reynolds-number dataset (Phase 4)

### Thus, validation is used **only for model selection**, not for Reynolds-number generalization.

---

### Split strategy adopted

### For each Reynolds-number dataset used in training, the data is split **independently** into training and validation subsets:

### $$[\mathcal{D}(Re_\lambda) = \mathcal{D}*{\text{train}}(Re*\lambda) \cup \mathcal{D}*{\text{val}}(Re*\lambda)]$$

### with:

### $$[\frac{|\mathcal{D}*{\text{val}}(Re*\lambda)|}{|\mathcal{D}(Re_\lambda)|} = 0.3]$$

### $$[\frac{|\mathcal{D}*{\text{train}}(Re*\lambda)|}{|\mathcal{D}(Re_\lambda)|} = 0.7]$$

### The split is **random** and performed along the wall-normal samples.

---

### Justification for random splitting

### The DNS data is ordered in the wall-normal direction ( y^+ ).
### Using a contiguous (e.g. “last 30%”) split would introduce **systematic spatial bias**:

#### * Training data would over-represent near-wall regions
#### * Validation data would over-represent centerline regions
#### * Hyperparameters would be tuned for spatial extrapolation rather than learning quality

### To avoid this, **random sampling** is used so that both training and validation sets contain samples from:

#### * Viscous sublayer
#### * Buffer layer
#### * Log-layer
#### * Outer region

### This ensures:

#### * Balanced representation of turbulence physics
#### * Fair assessment of model generalization within the same flow regime
#### * Stable and meaningful validation metrics

---

### Concatenation across Reynolds numbers

### After splitting each Reynolds-number dataset individually, the resulting subsets are concatenated:

### $$[\mathcal{D}*{\text{train}}^{(c)} =\bigcup*{Re_\lambda \in \mathcal{R}*{\text{train}}^{(c)}}\mathcal{D}*{\text{train}}(Re_\lambda)]$$

### $$[\mathcal{D}*{\text{val}}^{(c)} =\bigcup*{Re_\lambda \in \mathcal{R}*{\text{train}}^{(c)}}\mathcal{D}*{\text{val}}(Re_\lambda)]$$
### where:

#### * ( c ) denotes the case index (1–4)
#### * $( \mathcal{R}_{\text{train}}^{(c)} )$ is the set of training Reynolds numbers for case ( c )

### This pooled dataset exposes the model to turbulence statistics across multiple Reynolds regimes while maintaining strict separation from the test Reynolds number.

---

### Role of the validation set

### The validation set is used to:

#### * Monitor training convergence (MLP)
#### * Tune hyperparameters (Optuna, grid/random search)
#### * Select the best-performing model configuration
#### * Prevent overfitting

### The validation set is **never** used for:

#### * Final performance reporting
#### * Reynolds-number generalization
#### * Model retraining

```python
df_550 = pd.read_csv("../Datasets/Re-550.csv")
df_1000 = pd.read_csv("../Datasets/Re-1000.csv")
df_2000 = pd.read_csv("../Datasets/Re-2000.csv")
df_5200 = pd.read_csv("../Datasets/Re-5200.csv")
```

```python
datasets = {"5200":df_5200,
            "2000":df_2000,
            "1000":df_1000,
            "550":df_550}
```

```python
features = ['b_12','dU/dy','epsilon','k']
```

```python
target = 'phi_12'
```

```python
def split_and_concatenate_by_case(
    datasets,        # dict: {Re_lambda: df}
    train_res,       # list of Re_lambda used for training in this case
    features,        # list of feature column names
    target,          # target column name (single Φ component)
    val_fraction=0.3,
    seed=42
):
    X_train_list, y_train_list = [], []
    X_val_list, y_val_list     = [], []

    for Re in train_res:
        df = datasets[Re]

        X = df[features]
        y = df[target]

        X_tr, X_val, y_tr, y_val = train_test_split(
            X, y,
            test_size=val_fraction,
            random_state=seed,
            shuffle=True
        )

        X_train_list.append(X_tr)
        y_train_list.append(y_tr)
        X_val_list.append(X_val)
        y_val_list.append(y_val)

    X_train = pd.concat(X_train_list, axis=0).reset_index(drop=True)
    y_train = pd.concat(y_train_list, axis=0).reset_index(drop=True)
    X_val   = pd.concat(X_val_list, axis=0).reset_index(drop=True)
    y_val   = pd.concat(y_val_list, axis=0).reset_index(drop=True)

    return X_train, X_val, y_train, y_val
```

```python
X_train_1, X_val_1, y_train_1, y_val_1 = split_and_concatenate_by_case(datasets,['550','1000','2000'],features,target,val_fraction=0.3,seed=42)
X_train_2, X_val_2, y_train_2, y_val_2 = split_and_concatenate_by_case(datasets,['550','1000','5200'],features,target,val_fraction=0.3,seed=42)
X_train_3, X_val_3, y_train_3, y_val_3 = split_and_concatenate_by_case(datasets,['550','5200','2000'],features,target,val_fraction=0.3,seed=42)
X_train_4, X_val_4, y_train_4, y_val_4 = split_and_concatenate_by_case(datasets,['5200','1000','2000'],features,target,val_fraction=0.3,seed=42)
```

## **3.4 Feature Normalization**

### Motivation for feature normalization

### The input feature set used in this study consists of heterogeneous turbulence quantities with vastly different physical units and numerical magnitudes:

#### * Reynolds stress anisotropy components $( b_{ij} ): ( \mathcal{O}(10^{-1}) )$
#### * Turbulence kinetic energy ( k ): $( \mathcal{O}(10^{-2} \text{–} 10^{0}) )$
#### * Dissipation rate $( \varepsilon )$: can vary by several orders of magnitude, especially near the wall
#### * Mean velocity gradient $( \partial U / \partial y )$: strongly peaked near the wall

### Without normalization, features with large numerical ranges would dominate the learning process, leading to:

#### * Poor conditioning of neural network optimization
#### * Biased tree splits in ensemble methods
#### * Slower convergence and unstable training
#### * Reduced interpretability of feature importance

### Feature normalization is therefore essential to ensure numerical stability and balanced learning.

---

### Normalization strategy adopted

### A **min–max scaling** strategy is employed to map each input feature independently to the interval:

### $$[x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}\quad \in [0, 1]]$$

### where:

#### * ( x ) denotes the original feature value
#### * $( x_{\min} ), ( x_{\max} )$ are computed **only from the training set**

###  This choice is motivated by:

#### * Stable gradient behavior for neural networks
#### * Bounded input space for tree-based models
#### * Preservation of monotonic feature relationships
#### * Simplicity and reproducibility

---

### Prevention of data leakage

### To maintain a clean separation between training and validation data, the normalization parameters are computed **exclusively** from the training set:

### $$[\text{Scaler} \leftarrow \text{fit}(\mathcal{D}_{\text{train}})]$$

### $$[\mathcal{D}*{\text{val}}' = \text{transform}(\mathcal{D}*{\text{val}})]$$

### No statistics from the validation or test datasets are used during scaling.
### This prevents inadvertent leakage of information across datasets and ensures honest validation metrics.

---

### Model-specific considerations

#### * **MLP (Neural Network)**
  #### Normalization is critical. Without scaling, gradient descent becomes unstable due to ill-conditioned loss landscapes.

#### * **Random Forest**
  #### While RFs are less sensitive to feature scaling, normalization improves numerical robustness and consistency across models.

#### * **Gradient Boosted Trees (XGBoost)**
  #### Scaling is not strictly required but ensures fair comparison with neural models and stabilizes hyperparameter optimization.

### For consistency and reproducibility, the same normalized feature set is used for all models.

---

### Physical interpretation after normalization

### Although normalization removes physical units from the feature space, it does **not** destroy physical meaning:

#### * Relative variations and trends across the wall-normal direction are preserved
#### * Inner-scaled physics remains encoded in feature distributions
#### * Dimensional consistency is restored implicitly through the learning target

### Thus, normalization is treated as a numerical preprocessing step rather than a physical transformation.

---

### Storage and reuse of normalization parameters

### The fitted scaler is saved alongside trained model parameters:

#### * Ensures consistent preprocessing during Phase 4
#### * Enables deployment in RANS solvers or surrogate models
#### * Guarantees reproducibility of results

---

```python
cases = {
    1: {
        'X_train': X_train_1,
        'X_val': X_val_1,
        'y_train': y_train_1,
        'y_val': y_val_1
    },
    2: {
        'X_train': X_train_2,
        'X_val': X_val_2,
        'y_train': y_train_2,
        'y_val': y_val_2
    },
    3: {
        'X_train': X_train_3,
        'X_val': X_val_3,
        'y_train': y_train_3,
        'y_val': y_val_3
    },
    4: {
        'X_train': X_train_4,
        'X_val': X_val_4,
        'y_train': y_train_4,
        'y_val': y_val_4
    }
}
```

```python
scaled_data = {}
scalers = {}

for case_id, case_data in cases.items():
    scaler = MinMaxScaler()

    X_train_scaled = scaler.fit_transform(case_data['X_train'])
    X_val_scaled   = scaler.transform(case_data['X_val'])

    scaled_data[case_id] = {
        'X_train': X_train_scaled,
        'X_val': X_val_scaled,
        'y_train': case_data['y_train'].values,
        'y_val': case_data['y_val'].values
    }

    scalers[case_id] = scaler
```

## **3.5 Models Used**

### Motivation for multi-model comparison

### The pressure-strain correlation represents a highly nonlinear tensor-valued closure with complex dependencies on turbulence anisotropy, mean strain, and dissipation. No single machine learning model can be assumed *a priori* to be optimal for this task.

### Therefore, three fundamentally different regression models are employed:

#### 1. **Multi-Layer Perceptron (MLP)** — parametric, gradient-based
#### 2. **Random Forest (RF)** — nonparametric, ensemble-based
#### 3. **Gradient Boosted Decision Trees (GBDT)** — additive boosting framework

### Using models with distinct inductive biases allows us to:

#### * Assess robustness of learned mappings
#### * Compare smooth function approximation vs rule-based partitioning
#### * Identify sensitivity to feature scaling and hyperparameters
#### * Benchmark against prior turbulence-ML literature

### Each model is trained independently for:

#### * Each **case** (1–4)
#### * **pressure–strain component** $( \Phi_{12} )$

---

## **3.6 Model 1: Multi-Layer Perceptron (MLP)**

### Why MLP?

### The MLP is a universal function approximator capable of representing complex nonlinear mappings between continuous variables. It is particularly suitable for modeling smooth tensorial relationships in turbulence closures.

### Advantages:

#### * Captures high-order nonlinear interactions
#### * Smooth functional representation
#### * Widely used in turbulence ML literature

### Limitations:

#### * Sensitive to feature scaling
#### * Requires careful monitoring of convergence

---

### Network architecture

### The MLP consists of a fully connected feedforward network:

#### * Input layer: number of neurons = number of features
#### * Hidden layers: identical width
#### * Activation: ReLU
#### * Output layer: single scalar $( \Phi_{ij} )$

### $$[\hat{\Phi}*{ij} = f*{\theta}(\mathbf{x})]$$

### where $( \mathbf{x} )$ is the normalized feature vector and $( \theta )$ denotes network parameters.

---

### Important hyperparameters

| Hyperparameter     | Role                            |
| ------------------ | ------------------------------- |
| Hidden layer width | Model capacity                  |
| Number of layers   | Depth of nonlinear interactions |
| Learning rate      | Optimization stability          |

---

### Training setup

#### * Loss function: Mean Squared Error (MSE)
#### * Optimizer: Adam
#### * Early stopping based on validation loss

---

### **Storing convergence history (required for Fig. 6)**

### We **must store training and validation losses per epoch**, per case.

### PyTorch training loop with loss storage



---

```python
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dim=10, n_hidden_layers=5):
        super().__init__()

        layers = []
        layers.append(nn.Linear(input_dim, hidden_dim))
        layers.append(nn.ReLU())

        for _ in range(n_hidden_layers - 1):
            layers.append(nn.Linear(hidden_dim, hidden_dim))
            layers.append(nn.ReLU())

        layers.append(nn.Linear(hidden_dim, 1))  

        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)
```

```python
def train_mlp(model_name, model, optimizer, criterion, X_train, y_train, X_val, y_val, n_epochs=100):
    train_losses = []
    val_losses   = []
    
    for epoch in range(n_epochs):
        model.train()          # Training
        optimizer.zero_grad()

        preds = model(X_train).squeeze()
        train_loss = criterion(preds, y_train)
        train_loss.backward()
        optimizer.step()

        model.eval()            # Validation
        with torch.no_grad():
            val_preds = model(X_val).squeeze()
            val_loss = criterion(val_preds, y_val)

        train_losses.append(train_loss.item())
        val_losses.append(val_loss.item())

    print(f"[{model_name}] Training complete")

    return {
        "model_name": model_name,
        "train_loss": train_losses,
        "val_loss": val_losses
    }
```

```python
cases = {
    1: {"X_train": scaled_data[1]["X_train"], "y_train": scaled_data[1]["y_train"], "X_val": scaled_data[1]["X_val"], "y_val": scaled_data[1]["y_val"]},
    2: {"X_train": scaled_data[2]["X_train"], "y_train": scaled_data[2]["y_train"], "X_val": scaled_data[2]["X_val"], "y_val": scaled_data[2]["y_val"]},
    3: {"X_train": scaled_data[3]["X_train"], "y_train": scaled_data[3]["y_train"], "X_val": scaled_data[3]["X_val"], "y_val": scaled_data[3]["y_val"]},
    4: {"X_train": scaled_data[4]["X_train"], "y_train": scaled_data[4]["y_train"], "X_val": scaled_data[4]["X_val"], "y_val": scaled_data[4]["y_val"]}
}
```

```python
mlp_histories = {}
mlp_models = {}
for case_id, data in cases.items():
    print(f"\n=== Training MLP | Case {case_id} ===")
    model = MLP(input_dim=data["X_train"].shape[1])
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    criterion = nn.MSELoss()
    history = train_mlp(
        model_name=f"MLP_Case_{case_id}",
        model=model,
        optimizer=optimizer,
        criterion=criterion,
        X_train=torch.FloatTensor(data["X_train"]),
        y_train=torch.FloatTensor(data["y_train"]),
        X_val=torch.FloatTensor(data["X_val"]),
        y_val=torch.FloatTensor(data["y_val"]),
        n_epochs=100
    )
    mlp_models[case_id] = model
    mlp_histories[case_id] = history
```

## **3.7 Model 2: Random Forest Regression**

### Why Random Forest?

### Random Forests provide a strong, interpretable baseline with minimal preprocessing requirements.

### Advantages:

#### * Naturally nonlinear
#### * Resistant to overfitting
#### * Provides feature importance
#### * No need for gradient-based optimization

### Limitations:

#### * Piecewise constant predictions
#### * Poor extrapolation beyond training data

---

### Important hyperparameters

| Hyperparameter | Role                   |
| -------------- | ---------------------- |
| `n_estimators` | Ensemble strength      |
| `max_depth`    | Bias–variance tradeoff |

---

```python
### Implementation
rf_models = {}
for case_id, data in cases.items():
    rf = RandomForestRegressor(
        n_estimators=5,
        max_depth=10,
        random_state=42,
        n_jobs=-1
    )
    rf.fit(data["X_train"], data["y_train"])
    rf_models[case_id] = rf
```

### Validation performance is computed but **no training history exists**, since RF is not iterative.

---

## **3.8 Model 3: Gradient Boosted Decision Trees (XGBoost)**

### Why GBDT?

### Gradient Boosted Trees combine the strengths of decision trees and gradient-based optimization. They are widely reported to perform best for turbulence regression tasks.

### Advantages:

#### * Excellent performance on structured data
#### * Fewer parameters than neural networks
#### * Stable and interpretable
#### * Strong prior success in turbulence ML

---

### Important hyperparameters

| Hyperparameter  | Role                   |
| --------------- | ---------------------- |
| `max_depth`     | Interaction complexity |
| `learning_rate` | Boosting stability     |
| `n_estimators`  | Model capacity         |

---

### Implementation

```python
gbdt_models = {}
for case_id, data in cases.items():
    xgb_model = xgb.XGBRegressor(
        objective='reg:squarederror',
        max_depth=10,
        learning_rate=0.05,
        n_estimators=5,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    xgb_model.fit(data["X_train"], data["y_train"])
    gbdt_models[case_id] = xgb_model
```

## **3.11 Bayesian Optimization of Gradient Boosted Decision Trees**

### **Motivation for Bayesian Optimization**

### Gradient Boosted Decision Trees (GBDT) exhibit strong performance in turbulence modeling due to their ability to:

#### * Capture nonlinear feature interactions
#### * Handle mixed-scale physical inputs
#### * Provide stable predictions in near-wall regions

### However, GBDT performance is **highly sensitive to hyperparameter selection**. Manual tuning or grid search is inefficient due to:

#### * High-dimensional hyperparameter space
#### * Strong coupling between parameters
#### * Expensive training costs

### Therefore, **Bayesian Optimization** is employed to systematically and efficiently identify optimal hyperparameters using validation performance.

---

### **Why Optuna?**

### Optuna is a modern Bayesian optimization framework that:

#### * Builds a probabilistic surrogate model of the objective function
#### * Selects new hyperparameter configurations based on past evaluations
#### * Efficiently balances exploration and exploitation

### Compared to grid or random search, Optuna converges to high-performing configurations with **significantly fewer trials**.

---

### **Optimization Objective**

### For each training case and each pressure–strain component $( \Phi_{ij} )$, Bayesian optimization is performed to minimize the **validation mean squared error (MSE)**:

### $$[\mathcal{L}*{\text{val}} = \frac{1}{N}\sum*{n=1}^{N}\left(\Phi_{ij}^{(n)} - \hat{\Phi}_{ij}^{(n)}\right)^2]$$

### where:

#### * $( \Phi_{ij} )$ are DNS targets
#### * $( \hat{\Phi}_{ij} )$ are GBDT predictions
#### * Validation data is **never used for training**

---

### **Hyperparameters Tuned**

### The following GBDT hyperparameters are optimized:

| Hyperparameter     | Physical / ML Interpretation              |
| ------------------ | ----------------------------------------- |
| `max_depth`        | Controls nonlinear interaction complexity |
| `learning_rate`    | Governs boosting step size and stability  |
| `n_estimators`     | Number of boosting stages                 |
| `subsample`        | Controls sample-level regularization      |
| `colsample_bytree` | Feature-level regularization              |

### These parameters collectively control the **bias–variance tradeoff** and **generalization behavior**.

---

### **Bayesian Optimization Workflow**

### For each **case $( c \in {1,2,3,4} )$** and **component $( \Phi_{ij} )$**:

#### 1. **Fix training and validation datasets**
#### 2. **Define Optuna objective function**
#### 3. **Train a temporary GBDT model** using trial parameters
#### 4. **Evaluate validation MSE**
#### 5. **Update Optuna’s surrogate model**
#### 6. **Repeat for multiple trials**

### No test data is accessed during this process.

---

### **Optuna Objective Function**

```python
def create_objective(X_train, y_train, X_val, y_val):
    def objective(trial):
        params = {
            "max_depth": trial.suggest_int("max_depth", 6, 100),
            "learning_rate": trial.suggest_float("learning_rate", 1e-3, 1e-1, log=True),
            "n_estimators": trial.suggest_int("n_estimators", 5, 150),
            "subsample": trial.suggest_float("subsample", 0.6, 1.0),
            "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
            "objective": "reg:squarederror",
            "random_state": 42
        }
        model = xgb.XGBRegressor(**params)
        model.fit(X_train, y_train)
        preds = model.predict(X_val)
        mse = np.mean((preds - y_val) ** 2)
        return mse
    return objective
```

```python
best_params_dict = {}
best_val_mse_dict = {}
for case_id, data in cases.items():
    objective = create_objective(scaled_data[case_id]["X_train"], scaled_data[case_id]["y_train"], scaled_data[case_id]["X_val"], scaled_data[case_id]["y_val"])
    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=50)
    
    best_params = study.best_params
    best_val_mse = study.best_value
    best_params_dict[case_id] = best_params
    best_val_mse_dict[case_id] = best_val_mse 
```

## **Application of Optuna Results (Critical Step)**

### **Important Clarification**

### > **The model trained during Optuna trials is NOT the final model.**

### Each Optuna trial trains a **temporary GBDT** whose only purpose is to:

#### * Evaluate a candidate hyperparameter set
#### * Estimate validation performance

### These trial models are **discarded** after evaluation.

---

### **Training the Final Optimized GBDT Model**

### Once Bayesian optimization is complete:

#### 1. Extract the **best hyperparameters**
#### 2. Instantiate a **new GBDT model**
#### 3. Train it from scratch using:

   #### * The same training dataset
   #### * The optimized hyperparameters

```python
optimized_gbdt_models = {}
for case_id, data in cases.items():
    optimized_gbdt = xgb.XGBRegressor(
        **best_params_dict[case_id],
        objective="reg:squarederror",
        random_state=42
    )
    
    optimized_gbdt.fit(data["X_train"], data["y_train"])
    optimized_gbdt_models[case_id] = optimized_gbdt
    print(f"Optimized hyperparameters for Case_{case_id} are :",best_params_dict[case_id])
```

---

### **Why Retrain After Optimization?**

### Retraining is necessary because:

#### * Trial models may terminate early
#### * Internal training states differ across trials
#### * Clean reproducibility is required

### This ensures the final model is:

 #### - Trained exactly once
 #### - Using the optimal hyperparameters
 #### - Free from optimization artifacts

---

## **Role of Validation Set in Bayesian Optimization**

### The validation set serves three critical purposes:

#### 1. **Guides hyperparameter selection**
#### 2. **Prevents overfitting to training data**
#### 3. **Approximates generalization behavior within known Reynolds numbers**

### Validation data is:

#### * Never used for training
#### * Never reused in Phase 4 testing

## Phase 3 Conclusion

### In this phase, physics-informed machine learning models were trained to predict the pressure–strain correlation component $(\phi_{12})$ for turbulent channel flow using DNS data. The learning problem was formulated as a scalar regression task, with input features derived from turbulence modeling theory, including Reynolds stress anisotropy, turbulence kinetic energy, dissipation rate, and the mean velocity gradient. A Reynolds-number-aware data organization was strictly enforced to avoid information leakage and to preserve physical consistency.

### Three regression models, namely a multilayer perceptron, a random forest, and a gradient boosted decision tree, were trained and validated across four independent case configurations. For each case, training and validation were performed using data from three Reynolds numbers, while the remaining Reynolds number was completely excluded from this phase. Bayesian optimization using Optuna was applied to tune the hyperparameters of the gradient boosted decision tree model based solely on validation performance. All trained models and preprocessing pipelines were serialized for reproducibility and subsequent evaluation.

### The outcomes of Phase 3 demonstrate that machine learning models can successfully learn the nonlinear relationship between physically meaningful input features and the pressure–strain component $(\phi_{12})$ without prior exposure to unseen Reynolds numbers. This phase establishes a stable and physically consistent training framework, preparing the models for rigorous generalization assessment.

---

## Transition to Phase 4

### While Phase 3 focused on training and validation within known Reynolds-number regimes, the ultimate objective of this study is to evaluate whether the learned models can generalize to unseen flow conditions. Phase 4 addresses this objective by testing the trained models on Reynolds-number datasets that were entirely excluded during training and hyperparameter tuning.

### In Phase 4, each trained model will be evaluated on the held-out Reynolds number corresponding to its case configuration. Model performance will be assessed using quantitative metrics such as the coefficient of determination $((R^2))$, along with comparisons between predicted and DNS profiles of $(\phi_{12})$ across the wall-normal direction. This phase will provide direct insight into the Reynolds-number generalization capability of physics-informed machine learning models for pressure-strain closure.
