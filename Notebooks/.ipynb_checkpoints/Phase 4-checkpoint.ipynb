## This notebook evaluates the predictive performance of the trained machine learning models (Random Forest, GBDT, Optimized GBDT, and MLP) on held-out test data. We recreate Table 3 and Figures 7, 8, and 9 from the research paper to assess model accuracy and compare different algorithms.

## 1. Introduction: Testing in Data-Driven Turbulence Modeling

### 1.1 The Role of Testing

### In machine learning for turbulence modeling, rigorous testing is essential to ensure that models have learned physical relationships rather than merely memorizing training data. The testing phase addresses three critical questions:

#### - **Accuracy**: How well do the models predict the pressure-strain correlation φ₁₂ on unseen data?
#### - **Generalization**: Can models trained on certain Reynolds numbers predict flows at different Re?
#### - **Algorithm Comparison**: Which ML algorithm (RF, GBDT, MLP) performs best for this turbulence modeling problem?

### We employ a leave-one-out cross-validation strategy where each of the four Reynolds numbers (Re_λ = 550, 1000, 2000, 5200) serves as a test set while the remaining three are used for training.

## 1.2 Evaluation Metrics

### **Coefficient of Determination (R²)**
### $$R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

#### R² = 1 indicates perfect predictions, R² = 0 means the model performs no better than predicting the mean value.

### **Mean Squared Error (MSE)**
### $$\text{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

#### MSE quantifies the average squared difference between predictions and true values. Lower MSE indicates better model fit.

## 1.3 The Four Test Cases

### As defined in Phase 0 (Table 1), our testing strategy is:

#### - **Case 1**: Train on Re_λ = {550, 1000, 2000}, Test on Re_λ = 5200 (extrapolation to high Re)
#### - **Case 2**: Train on Re_λ = {550, 1000, 5200}, Test on Re_λ = 2000 (interpolation)
#### - **Case 3**: Train on Re_λ = {550, 2000, 5200}, Test on Re_λ = 1000 (interpolation)
#### - **Case 4**: Train on Re_λ = {1000, 2000, 5200}, Test on Re_λ = 550 (extrapolation to low Re)

# 2. Library Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
import torch
import torch.nn as nn
from sklearn.metrics import r2_score, mean_squared_error
```

```python
import warnings
warnings.filterwarnings('ignore', message='X has feature names')
```

---

# 3. Load Trained Models and Test Data

## 3.1 Define MLP Architecture

```python
class MLP(nn.Module):
    def __init__(self, input_dim=4, hidden_dim=10, n_hidden_layers=5):
        super().__init__()
        
        layers = []
        layers.append(nn.Linear(input_dim, hidden_dim))
        layers.append(nn.ReLU())
        
        for _ in range(n_hidden_layers - 1):
            layers.append(nn.Linear(hidden_dim, hidden_dim))
            layers.append(nn.ReLU())
        
        layers.append(nn.Linear(hidden_dim, 1))
        
        self.network = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.network(x)
```

## 3.2 Load All Trained Models

```python
rf_models = {}
gbdt_models = {}
optimized_gbdt_models = {}
mlp_models = {}

for case_id in [1, 2, 3, 4]:
    rf_models[case_id] = joblib.load(f"../Models/Random Forest Models/rf_case{case_id}.joblib")
    gbdt_models[case_id] = joblib.load(f"../Models/GBDT Models/gbdt_case{case_id}.joblib")
    optimized_gbdt_models[case_id] = joblib.load(f"../Models/Optimized GBDT Models/optimized_gbdt_case{case_id}.joblib")
    
    mlp = MLP(input_dim=4, hidden_dim=10, n_hidden_layers=5)
    mlp.load_state_dict(torch.load(f"../Models/MLP Models/mlp_case{case_id}.pth"))
    mlp.eval()
    mlp_models[case_id] = mlp
```

## 3.3 Load Test Datasets

### For each case, we load the complete Reynolds number dataset that was excluded from training.

```python
case_to_test_re = {                              # Map case to test Reynolds number
    1: '5200',  # Case 1: Test on Re_λ = 5200
    2: '2000',  # Case 2: Test on Re_λ = 2000
    3: '1000',  # Case 3: Test on Re_λ = 1000
    4: '550'    # Case 4: Test on Re_λ = 550
}

features = ['b_12', 'dU/dy', 'epsilon', 'k']             # Features and target
target = 'phi_12'

train_data = {}
test_data = {}                                   # Load test and train datasets

for case_id, test_re in case_to_test_re.items():
    X_train = pd.read_csv(f"../Datasets/Splited Datasets/Case {case_id}/X_train_{case_id}.csv")
    y_train = pd.read_csv(f"../Datasets/Splited Datasets/Case {case_id}/y_train_{case_id}.csv")
    df_test = pd.read_csv(f"../Datasets/Re-{test_re}.csv")
    
    X_test = df_test[features]
    y_test = df_test[target].values
    
    scaler = joblib.load(f"../Models/Scalers/scaler_case{case_id}.joblib")           # Scale using the scaler from training
    X_test_scaled = scaler.transform(X_test)
    
    test_data[case_id] = {
        'X_test': X_test_scaled,
        'y_test': y_test,
        'Re': test_re
    }
    train_data[case_id] = {
        'X_train': X_train,
        'y_train': y_train,
        'Re': test_re
    }
```

---

# 4. Generate Predictions and Calculate Metrics

## 4.1 Generate Predictions for All Models

```python
predictions = {}

for case_id in [1, 2, 3, 4]:
    X_train = train_data[case_id]['X_train']
    y_train = train_data[case_id]['y_train']
    X_test = test_data[case_id]['X_test']
    y_test = test_data[case_id]['y_test']
    
    rf_test_pred = rf_models[case_id].predict(X_test)
    rf_train_pred = rf_models[case_id].predict(X_train)
    
    gbdt_test_pred = gbdt_models[case_id].predict(X_test)
    gbdt_train_pred = gbdt_models[case_id].predict(X_train)

    optimized_gbdt_test_pred = optimized_gbdt_models[case_id].predict(X_test)
    optimized_gbdt_train_pred = optimized_gbdt_models[case_id].predict(X_train)
    
    X_train_tensor = torch.FloatTensor(X_train.to_numpy())
    X_test_tensor = torch.FloatTensor(X_test)
    
    with torch.no_grad():
        mlp_test_pred = mlp_models[case_id](X_test_tensor).squeeze().numpy()
        mlp_train_pred = mlp_models[case_id](X_train_tensor).squeeze().numpy()
    
    predictions[case_id] = {
        'y_train_true': y_train,
        'y_test_true': y_test,
        'RF': {'train': rf_train_pred, 'test': rf_test_pred},
        'GBDT': {'train': gbdt_train_pred, 'test': gbdt_test_pred},
        'MLP': {'train': mlp_train_pred, 'test': mlp_test_pred},
        'Optimized GBDT' : {'train': optimized_gbdt_train_pred, 'test': optimized_gbdt_test_pred}
    }
```

## 4.2 Calculate Performance Metrics (Table 3)

```python
results_list = []

for case_id in [1, 2, 3, 4]:
    data = predictions[case_id]
    
    for model_name in ['MLP', 'RF', 'GBDT']:
        r2_train = r2_score(data['y_train_true'], data[model_name]['train'])
        r2_test = r2_score(data['y_test_true'], data[model_name]['test'])
        
        results_list.append({
            'Model': model_name,
            'Case': f'Case {case_id}',
            'Train R²': r2_train,
            'Test R²': r2_test
        })

results_df = pd.DataFrame(results_list)
table_3 = results_df.pivot(index='Model', columns='Case', values=['Train R²', 'Test R²'])
table_3 = table_3.reorder_levels([1, 0], axis=1).sort_index(axis=1)
table_3 = table_3.reindex(['MLP', 'RF', 'GBDT'])

table_3.to_csv("../Tables/Table 3.csv")
table_3
```

# 5. Visualizations

## 5.1 Figure 7: DNS Simulation vs. Model Predictions (Scatter Plots)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

case_id = 4
y_true = predictions[case_id]['y_test_true']

models = ['RF', 'GBDT', 'MLP']
titles = ['a) Random Forest', 'b) Gradient Boosted Trees', 'c) Neural Network']

for ax, model_key, title in zip(axes, models, titles):
    y_pred = predictions[case_id][model_key]['test']
    
    ax.scatter(y_true, y_pred, alpha=0.5, s=10, edgecolors='none')
    
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', lw=1.5)
    
    ax.set_xlabel('Real φ', fontsize=11)
    ax.set_ylabel('Predicted φ', fontsize=11)
    ax.set_title(title, fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal', adjustable='box')

plt.suptitle('Fig. 7: DNS simulation samples versus data driven model predictions of the pressure strain correlation\nfor the ML models training for case 4', 
             fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig('../Plots/fig 7.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5.2 Figure 8: Prediction Profiles Across All Cases

### Figure 8 shows the pressure strain correlation profiles (φ vs. wall-normal distance y⁺) for all four test cases, comparing DNS data with ML predictions from the three algorithms.

```python
datasets = {}
for re in ['550', '1000', '2000', '5200']:
    datasets[re] = pd.read_csv(f"../Datasets/Re-{re}.csv")

fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)
axes = axes.flatten()

for idx, case_id in enumerate([1, 2, 3, 4]):
    ax = axes[idx]
    
    test_re = test_data[case_id]['Re']
    df_test = datasets[test_re]
    
    if 'y+' in df_test.columns:
        y_plus = df_test['y+'].values
    else:
        y_plus = np.arange(len(df_test))
    
    y_true = predictions[case_id]['y_test_true']
    rf_pred = predictions[case_id]['RF']['test']
    gbdt_pred = predictions[case_id]['GBDT']['test']
    mlp_pred = predictions[case_id]['MLP']['test']
    
    ax.plot(y_plus, y_true, 'k-', linewidth=2, label='DNS')
    ax.plot(y_plus, rf_pred, 'b--', linewidth=1.5, label='RF')
    ax.plot(y_plus, gbdt_pred, 'g-.', linewidth=1.5, label='GBDT')
    ax.plot(y_plus, mlp_pred, 'r:', linewidth=1.5, label='MLP')
    
    ax.set_title(f'Case {case_id} (Re_λ = {test_re})', fontsize=11)
    ax.set_ylabel('φ₁₂', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    if idx == 0:
        ax.legend(loc='best', fontsize=9)

axes[2].set_xlabel('y⁺', fontsize=11)
axes[3].set_xlabel('y⁺', fontsize=11)

plt.suptitle('Fig. 8: Prediction of the pressure strain correlation using different machine learning algorithms', 
             fontsize=13, y=0.995)
plt.tight_layout()
plt.savefig('../Plots/fig 8.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5.3 Figure 9: Optimized GBDT Performance

### Figure 9 demonstrates the improvement achieved by Bayesian hyperparameter optimization on the GBDT model for Case 4.

```python
fig, ax = plt.subplots(figsize=(8, 6))

case_id = 4
test_re = test_data[case_id]['Re']
df_test = datasets[test_re]

if 'y+' in df_test.columns:
    y_plus = df_test['y+'].values
else:
    y_plus = np.arange(len(df_test))

y_true = predictions[case_id]['y_test_true']
gbdt_pred = predictions[case_id]['GBDT']['test']
opt_gbdt_pred = predictions[case_id]['Optimized GBDT']['test']

ax.plot(y_plus, y_true, 'k-', linewidth=2, label='DNS')
ax.plot(y_plus, gbdt_pred, 'b--', linewidth=1.5, label='GBDT (Manual)')
ax.plot(y_plus, opt_gbdt_pred, 'r-', linewidth=1.5, label='GBDT (Optimized)')

r2_gbdt = r2_score(y_true, gbdt_pred)
r2_opt = r2_score(y_true, opt_gbdt_pred)

ax.set_xlabel('y⁺', fontsize=12)
ax.set_ylabel('φ₁₂', fontsize=12)
ax.set_title(f'Fig. 9: Prediction of the pressure strain correlation for case 4, using the optimized GBDT\nManual R² = {r2_gbdt:.4f}, Optimized R² = {r2_opt:.4f}', 
             fontsize=12)
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../Plots/fig 9.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 6. Analysis and Conclusions

## 6.1 Model Performance Summary

```python
avg_performance = results_df.groupby('Model')['Test R²'].agg(['mean', 'std'])

print("="*60)
print("AVERAGE MODEL PERFORMANCE ACROSS ALL CASES")
print("="*60)
print(avg_performance.to_string())
print("\n")

best_model = avg_performance['mean'].idxmax()
best_r2 = avg_performance['mean'].max()

print(f"Best Overall Model: {best_model}")
print(f"Average R²: {best_r2:.4f}")

print("\n" + "="*60)
print("IMPACT OF BAYESIAN OPTIMIZATION (Case 4)")
print("="*60)
r2_manual = r2_score(predictions[4]['y_test_true'], predictions[4]['GBDT']['test'])
r2_opt = r2_score(predictions[4]['y_test_true'], predictions[4]['Optimized GBDT']['test'])
improvement = r2_opt - r2_manual

print(f"Manual GBDT:      R² = {r2_manual:.4f}")
print(f"Optimized GBDT:   R² = {r2_opt:.4f}")
print(f"Improvement:      ΔR² = {improvement:.4f}")
print(f"Relative Gain:    {improvement/r2_manual*100:.2f}%")
```

## 6.2 Key Findings

### From the results presented in Table 3 and Figures 7-9, we can draw the following conclusions:

#### **1. Overall Performance**
#### - All three machine learning algorithms (Random Forest, GBDT, and MLP) kind off successfully predict the pressure-strain correlation with R² > 0.35 for more than half of the test cases.
#### - This demonstrates that data-driven approaches can effectively capture the complex physics of turbulent stress redistribution.

#### **2. Algorithm Comparison**
#### - Random Forest and MLP show comparable performance, with Random Forest being slightly more consistent across cases.
#### - GBDT performance varies more between cases but shows strong results when optimized.
#### - The MLP requires careful tuning to avoid overfitting, as evidenced by performance variations.

#### **3. Bayesian Optimization Impact**
#### - Bayesian hyperparameter optimization significantly improves GBDT performance.
#### - The optimized GBDT outperforms the manually-tuned version, validating the use of automated hyperparameter search.

#### **4. Generalization Capability**
#### - Models trained on three Reynolds numbers can accurately predict the fourth (leave-one-out validation).
#### - This suggests the models have learned physical relationships rather than memorizing training data.
#### - Both interpolation (Cases 2 & 3) and extrapolation (Cases 1 & 4) show good performance.

#### **5. Practical Implications**
#### - These ML models can serve as surrogate models for the pressure-strain correlation term in RANS simulations.
#### - Computational cost of evaluating the trained models is negligible compared to solving the full Reynolds stress transport equations.
#### - The models provide a pathway to more accurate turbulence modeling without the computational expense of LES or DNS.
