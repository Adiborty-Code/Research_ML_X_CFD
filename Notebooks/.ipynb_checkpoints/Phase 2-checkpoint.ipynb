## Analysis 1 : Dataset Sanity & Physical Consistency Checks

### Before applying any machine learning model, it is essential to verify that the constructed datasets
### are physically meaningful and consistent with known turbulent channel flow behavior.

### The objective of this section is to ensure that:

#### - The datasets exhibit correct wall-normal structure
#### - No numerical or preprocessing artifacts are present
#### - The datasets behave consistently across Reynolds numbers
#### - The data represents realistic turbulent statistics

### This step is critical to avoid training models on corrupted or non-physical data.

### -> Dataset Size Across Reynolds Numbers

### Each dataset corresponds to a different friction Reynolds number and may contain a different number of wall-normal sampling points.
### Comparing dataset sizes helps assess grid resolution consistency.

```python
dataset_sizes = pd.DataFrame({               # DataFrame showing all corresponding size of the four datasets
    "Re_lambda": [550, 1000, 2000, 5200],
    "Number of points": [
        len(df_550),
        len(df_1000),
        len(df_2000),
        len(df_5200)
    ]
})

dataset_sizes
```

### -> Grid Resolution vs Reynolds Number

### This plot illustrates how the wall-normal grid resolution increases with Reynolds number, as expected for DNS datasets.

```python
# Plot showing all the wall normal points for all the four Re(s)
plt.figure()
plt.plot(dataset_sizes["Re_lambda"],dataset_sizes["Number of points"], marker='o')
plt.xlabel(r"$Re_\lambda$")
plt.ylabel("Number of wall-normal points")
plt.title("Wall-normal grid resolution across Reynolds numbers")
plt.grid(True)
plt.show()
```

### -> Monotonicity of Wall-normal Coordinate

### The wall-normal coordinate $y^+$ must be strictly increasing.
### Any violation would indicate a corrupted dataset or incorrect preprocessing.

```python
def check_monotonicity(df, name):                      # Function for checking monotonicity of the datasets
    is_monotonic = df["y^+"].is_monotonic_increasing
    print(f"{name}: y⁺ monotonic increasing → {is_monotonic}")

check_monotonicity(df_550, "Re_550")
check_monotonicity(df_1000, "Re_1000")
check_monotonicity(df_2000, "Re_2000")
check_monotonicity(df_5200, "Re_5200")
```

### -> Missing and Invalid Values Check

### Machine learning models are sensitive to missing or infinite values.
### This check ensures numerical cleanliness of the dataset.

```python
def check_nan_inf(df, name):             # Function for checking all the missing and invalid values and printing out the respective counts
    nan_count = df.isna().sum().sum()
    inf_count = np.isinf(df.values).sum()
    print(f"{name}: NaNs = {nan_count}, Infs = {inf_count}")

check_nan_inf(df_550, "Re_550")
check_nan_inf(df_1000, "Re_1000")
check_nan_inf(df_2000, "Re_2000")
check_nan_inf(df_5200, "Re_5200")
```

### -> Wall-normal Grid Spacing Comparison

### DNS grids are non-uniform in the wall-normal direction, with finer spacing near the wall.
### Comparing $\\Delta y^+$ distributions verifies correct grid clustering.

```python
def compute_dy_plus(df):               # Function for calculating delta y^+
    return df["y^+"].diff().dropna()

dy_550 = compute_dy_plus(df_550)
dy_1000 = compute_dy_plus(df_1000)
dy_2000 = compute_dy_plus(df_2000)
dy_5200 = compute_dy_plus(df_5200)
```

```python
plt.figure()                                                    # Code for plotting it
plt.plot(df_550["y^+"][1:], dy_550, label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"][1:], dy_1000, label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"][1:], dy_2000, label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"][1:], dy_5200, label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\Delta y^+$")
plt.yscale("log")
plt.legend()
plt.title("Wall-normal grid spacing comparison")
plt.grid(True)
plt.show()
```

### -> Near-wall and Centerline Statistics

### To verify physical realism, representative quantities are compared near the wall and near the channel centerline.
### Large gradients near the wall and smoother behavior at the centerline are characteristic of turbulent channel flow.

```python
def wall_center_stats(df, column):                 # Function for finding the wall mean and centerline mean for any df and any column
    wall = df[df["y^+"] < 10][column]
    center = df[df["y^+"] > df["y^+"].max() * 0.9][column]
    
    return {
        "Wall mean": wall.mean(),
        "Centerline mean": center.mean()
    }

columns = ["k", "epsilon", "b_12", "u'v'"]

for col in columns:
    print(col, wall_center_stats(df_5200, col))
```

```python
columns = ["k", "epsilon", "b_12", "u'v'"]

for col in columns:
    print(col, wall_center_stats(df_2000, col))
```

```python
columns = ["k", "epsilon", "b_12", "u'v'"]

for col in columns:
    print(col, wall_center_stats(df_1000, col))
```

```python
columns = ["k", "epsilon", "b_12", "u'v'"]

for col in columns:
    print(col, wall_center_stats(df_550, col))
```

### Summary of Sanity Checks

#### - All datasets exhibit strictly increasing $y^+$
#### - No missing or infinite values are present
#### - Grid resolution increases with Reynolds number
#### - Wall-normal spacing shows expected near-wall clustering
#### - Physical quantities exhibit realistic wall and centerline behavior

### These checks confirm that the datasets are physically consistent and suitable for turbulence modeling using machine learning.

## Analysis 2 : Mean Flow Physics

### Before analyzing turbulent fluctuations or applying machine learning models, it is essential to understand the behavior of the mean flow.
### In a fully developed turbulent channel flow, the mean velocity field provides the primary source of turbulence production.

### In this section, we analyze:

#### - The mean streamwise velocity profile, $U(y)$
#### - The mean velocity gradient (shear), $\mathrm{d}U/\mathrm{d}y$

### These quantities govern turbulence production and strongly influence the structure of Reynolds stresses and pressure-strain correlations.

### -> Mean Streamwise Velocity Profile

### For fully developed plane channel flow, the mean velocity has only one non-zero component in the streamwise direction:

### $$ \langle \mathbf{U} \rangle = \left( U(y),\; 0,\; 0 \right) $$

### This is a consequence of statistical homogeneity in the streamwise and spanwise directions, and symmetry about the channel centerline.

### The velocity is presented in inner (wall) units:

### $$ U^+ = \frac{U}{u_\tau}, \qquad y^+ = \frac{y u_\tau}{\nu} $$

```python
plt.figure()
plt.semilogx(df_550["y^+"], df_550["U"], label=r"$Re_\lambda=550$")
plt.semilogx(df_1000["y^+"], df_1000["U"], label=r"$Re_\lambda=1000$")
plt.semilogx(df_2000["y^+"], df_2000["U"], label=r"$Re_\lambda=2000$")
plt.semilogx(df_5200["y^+"], df_5200["U"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$U^+$")
plt.title("Mean streamwise velocity profile")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Physical Interpretation

#### - **Viscous sublayer** ($y^+ \lesssim 5$):  
  #### The velocity profile is approximately linear, $U^+ \approx y^+$, indicating dominance of viscous stresses.

#### - **Buffer layer** ($5 \lesssim y^+ \lesssim 30$):  
  #### Transition region where both viscous and turbulent stresses are important.

#### - **Logarithmic region** ($y^+ \gtrsim 30$):  
  #### The velocity follows the log-law,
  #### $$ U^+ = \frac{1}{\kappa} \ln y^+ + B $$
  #### where $\kappa$ is the von Kármán constant.

#### - **Reynolds number effect**:  
  #### Higher $Re_\lambda$ extends the logarithmic region, indicating stronger scale separation.

### -> Mean Velocity Gradient

### The wall-normal gradient of the mean velocity, $\mathrm{d}U/\mathrm{d}y$, represents the mean strain rate:

### $$ S_{12} = \frac{1}{2} \frac{\mathrm{d}U}{\mathrm{d}y} $$

### This quantity plays a central role in turbulence production, as it couples with the Reynolds shear stress $\langle u'v' \rangle$.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["dU/dy"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["dU/dy"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["dU/dy"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["dU/dy"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\mathrm{d}U/\mathrm{d}y$")
plt.title("Mean velocity gradient across Reynolds numbers")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Physical Interpretation

#### - The mean shear is strongest near the wall, where velocity gradients are largest.
#### - Moving away from the wall, $\mathrm{d}U/\mathrm{d}y$ decreases rapidly.
#### - The near-wall region dominates turbulence production via
  ### $$ P = -\langle u'v' \rangle \frac{\mathrm{d}U}{\mathrm{d}y} $$

### This explains why accurate modeling of pressure-strain correlations requires precise representation of the mean velocity gradient.

### -> Why Only One Mean Velocity Component Exists

### In statistically steady, fully developed channel flow:

#### - There is no mean pressure gradient in the wall-normal or spanwise directions
#### - The flow is homogeneous in $x$ and $z$
#### - No mean mass flux exists in the $y$ or $z$ directions

### As a result:

### $$ \langle V \rangle = \langle W \rangle = 0 $$

### and the mean velocity field reduces to:

### $$ \langle \mathbf{U} \rangle = (U(y), 0, 0) $$

### This physical constraint explains why the feature space contains only a single mean velocity component.

### Summary

#### - The mean velocity profile exhibits classical turbulent channel flow behavior
#### - Mean shear is concentrated near the wall and decreases away from it
#### - Only one mean velocity component exists due to flow symmetry and homogeneity
#### - These properties directly influence turbulence production and anisotropy

### Understanding the mean flow is essential before analyzing Reynolds stresses and pressure-strain correlations.

## Analysis 3 : Reynolds Stress Tensor Behavior

### The Reynolds stress tensor characterizes the intensity and anisotropy of turbulent velocity fluctuations.
### Its structure plays a central role in turbulence modeling and highlights the limitations of linear eddy-viscosity models (EVMs).

### The Reynolds stress tensor is defined as:

### $$ R_{ij} = \langle u_i' u_j' \rangle $$

### For turbulent channel flow, the dominant non-zero components are:

### $$ R_{11} = \langle u'u' \rangle, \quad R_{22} = \langle v'v' \rangle, \quad R_{33} = \langle w'w' \rangle, \quad R_{12} = \langle u'v' \rangle $$

### These components are examined as functions of the wall-normal coordinate $y^+$.

### -> Streamwise Normal Stress: $\langle u'u' \rangle$

### The streamwise velocity fluctuations dominate turbulent kinetic energy due to strong mean shear and streak formation near the wall.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["u'u'"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["u'u'"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["u'u'"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["u'u'"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\langle u'u' \rangle$")
plt.title("Streamwise Reynolds stress")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Wall-normal Normal Stress: $\langle v'v' \rangle$

### Wall-normal fluctuations are suppressed near the wall due to the impermeability condition, leading to reduced $R_{22}$ values.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["v'v'"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["v'v'"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["v'v'"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["v'v'"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\langle v'v' \rangle$")
plt.title("Wall-normal Reynolds stress")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Spanwise Normal Stress: $\langle w'w' \rangle$

### Spanwise fluctuations are intermediate in magnitude, reflecting redistribution of turbulent energy by pressure-strain effects.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["w'w'"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["w'w'"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["w'w'"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["w'w'"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\langle w'w' \rangle$")
plt.title("Spanwise Reynolds stress")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Reynolds Shear Stress: $\langle u'v' \rangle$

### The Reynolds shear stress represents turbulent momentum transport and directly couples with the mean velocity gradient.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["u'v'"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["u'v'"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["u'v'"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["u'v'"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\langle u'v' \rangle$")
plt.title("Reynolds shear stress")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Physical Interpretation

#### - The streamwise normal stress $\langle u'u' \rangle$ dominates, reflecting the strong influence of mean shear and streak formation.

#### - The wall-normal component $\langle v'v' \rangle$ is suppressed near the wall due to the no-penetration boundary condition.

#### - The spanwise component $\langle w'w' \rangle$ lies between $R_{11}$ and $R_{22}$, indicating redistribution of energy by pressure-strain interactions.

#### - The Reynolds shear stress $\langle u'v' \rangle$ is negative throughout most of the channel, signifying turbulent momentum transport toward the wall.

#### - Increasing $Re_\lambda$ intensifies turbulence and extends the outer-layer influence, while inner-layer scaling remains approximately universal.

### -> Implications for Turbulence Modeling

### The strong inequality among Reynolds stress components demonstrates that turbulence is highly anisotropic.

### Linear eddy-viscosity models assume:

### $$ R_{ij} \propto S_{ij} $$

### which cannot represent the observed tensorial structure.

### This motivates the use of tensor-based, nonlinear modeling approaches, including machine learning methods capable of learning complex pressure-strain dynamics.

### Summary

#### - Reynolds stresses exhibit strong anisotropy across the channel
#### - Streamwise fluctuations dominate turbulent kinetic energy
#### - Reynolds shear stress governs momentum transfer
#### - Anisotropy persists across Reynolds numbers

### These observations highlight the necessity of tensor-aware machine learning models for turbulence closure.

## Analysis 4 : Turbulence Kinetic Energy

### Turbulence kinetic energy (TKE) represents the total kinetic energy associated with velocity fluctuations.
### It provides a scalar measure of turbulence intensity but does not contain information about anisotropy or directional energy distribution.

### TKE is defined as:

### $$ k = \frac{1}{2} \left( R_{11} + R_{22} + R_{33} \right) $$

### where

### $$ R_{11} = \langle u'u' \rangle, \quad R_{22} = \langle v'v' \rangle, \quad R_{33} = \langle w'w' \rangle$$

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["k"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["k"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["k"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["k"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$k$")
plt.title("Turbulence kinetic energy distribution")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Physical Interpretation

#### - Turbulence kinetic energy peaks in the near-wall region, where production due to mean shear is strongest.

#### - The peak location shifts slightly away from the wall as the Reynolds number increases, indicating enhanced outer-layer activity.

#### - Away from the wall, TKE decreases gradually toward the channel centerline, reflecting reduced turbulence production and increased energy redistribution.

#### - Increasing $Re_\lambda$ leads to higher overall turbulence intensity and a broader energetic region.

### -> Relation to Turbulence Production

### The production of turbulence kinetic energy is governed by:

### $$ P = - \langle u'v' \rangle \frac{\mathrm{d}U}{\mathrm{d}y} $$

### This explains why:

#### - High values of $k$ occur where both the Reynolds shear stress and mean velocity gradient are significant.

#### - Accurate modeling of $k$ requires correct representation of Reynolds stresses and mean shear.

### -> Limitations of Turbulence Kinetic Energy

### Although $k$ quantifies the magnitude of turbulent fluctuations, it does not provide information about:

#### - Directional distribution of turbulent energy
#### - Stress anisotropy
#### - Momentum transport mechanisms

### Different Reynolds stress states can share identical values of $k$.
### Therefore, scalar turbulence models based solely on $k$ are unable to represent complex anisotropic turbulence behavior.

### This motivates the inclusion of tensorial quantities, such as the Reynolds stress anisotropy tensor, in advanced turbulence models and machine learning approaches.

### Summary

#### - Turbulence kinetic energy measures the intensity of velocity fluctuations
#### - Peak $k$ occurs near the wall due to strong shear-driven production
#### - Higher Reynolds numbers increase turbulence intensity
#### - $k$ alone is insufficient to describe anisotropic turbulence

### These observations further justify the need for tensor-based modeling of turbulence closure terms.

## Analysis 5 : Turbulent Dissipation

### Turbulent dissipation represents the rate at which kinetic energy of velocity fluctuations is converted into internal energy through viscous effects.

### The scalar dissipation rate is defined as:

### $$ \varepsilon = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33} $$

### where $\varepsilon_{ij}$ are the viscous dissipation components associated with each velocity fluctuation direction.


```python
plt.figure()
plt.plot(df_550["y^+"], df_550["epsilon"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["epsilon"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["epsilon"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["epsilon"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\varepsilon$")
plt.title("Turbulent dissipation rate distribution")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Physical Interpretation

#### - The dissipation rate is strongly concentrated near the wall, where velocity gradients are largest.

#### - As $y^+$ increases, $\varepsilon$ decreases rapidly, reflecting reduced viscous action in the outer region.

#### - Higher Reynolds numbers exhibit larger peak dissipation, consistent with increased turbulence intensity and broader scale separation.

### -> Energy Cascade Interpretation

### In turbulent flows, kinetic energy is transferred from large-scale motions to progressively smaller scales through nonlinear interactions.
### This process continues until viscous effects dominate, resulting in dissipation.

### The dissipation rate $\varepsilon$ therefore represents the sink of turbulent kinetic energy in the cascade:

### $$ k \;\longrightarrow\; \text{inertial transfer} \;\longrightarrow\; \varepsilon $$

### -> Implications for Machine Learning Stability

### The dissipation rate sets a characteristic small-scale time scale:

### $$ \tau_\varepsilon \sim \frac{k}{\varepsilon} $$

### Regions with large $\varepsilon$ correspond to rapid dynamics and strong gradients, which can pose challenges for machine learning models.

### Including $\varepsilon$ as an input feature:

#### - Helps regularize predictions near the wall
#### - Provides scale-awareness to the model
#### - Improves robustness across Reynolds numbers

### This makes dissipation a critical quantity for normalization and model stability considerations.

### Summary

#### - Turbulent dissipation represents the viscous sink of kinetic energy
#### - Dissipation is strongest near the wall
#### - Higher Reynolds numbers increase dissipation magnitude
#### - Dissipation provides essential scale information for ML models

### This quantity plays a key role in both turbulence physics and the numerical stability of machine learning closures.

## Analysis 6 : Reynolds Stress Anisotropy Tensor

### The anisotropy of turbulence is quantified using the Reynolds stress anisotropy tensor, which measures deviations from isotropic turbulence.

### It is defined as:

### $$ b_{ij} = \frac{R_{ij}}{2k} - \frac{1}{3} \delta_{ij} $$

### where

#### - $R_{ij} = \langle u_i' u_j' \rangle$ is the Reynolds stress tensor
#### - $k = \frac{1}{2}(R_{11} + R_{22} + R_{33})$ is the turbulence kinetic energy
#### - $\delta_{ij}$ is the Kronecker delta

### The tensor $b_{ij}$ is traceless and encodes the directional distribution of turbulent kinetic energy.

### -> Normal Anisotropy Components

### The diagonal components of the anisotropy tensor represent the distribution of turbulent kinetic energy among the three velocity components:

### $$ b_{11}, \quad b_{22}, \quad b_{33} $$

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["b_11"], label=r"$b_{11},\ Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["b_11"], label=r"$b_{11},\ Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["b_11"], label=r"$b_{11},\ Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["b_11"], label=r"$b_{11},\ Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$b_{11}$")
plt.title("Streamwise anisotropy component")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Shear Anisotropy Component

### The off-diagonal anisotropy component $b_{12}$ represents anisotropy associated with turbulent momentum transport.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["b_12"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["b_12"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["b_12"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["b_12"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$b_{12}$")
plt.title("Shear anisotropy component")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Near-wall Anisotropy

### Near the wall, turbulence exhibits strong anisotropy due to boundary constraints:

#### - Wall-normal fluctuations are suppressed ($b_{22} < 0$)
#### - Streamwise fluctuations dominate ($b_{11} > 0$)
#### - Spanwise fluctuations remain intermediate

### This results in highly anisotropic turbulence states close to the wall, which cannot be represented by isotropic or linear models.

### -> Return to Isotropy

### Away from the wall, the anisotropy tensor components gradually decrease in magnitude, indicating a tendency toward isotropic turbulence:

### $$ b_{ij} \rightarrow 0 \quad \text{as} \quad y^+ \rightarrow \text{centerline} $$

### This return to isotropy is driven by pressure-strain interactions, which redistribute turbulent kinetic energy among velocity components.

### -> Why Pressure-Strain Exists Physically

### Pressure-strain correlation acts to:

#### - Redistribute turbulent kinetic energy
#### - Reduce anisotropy
#### - Drive turbulence toward isotropic states

### Mathematically, pressure-strain appears in the Reynolds stress transport equation as:

### $$ \Phi_{ij} = \left\langle \frac{p'}{\rho} \left( \frac{\partial u_i'}{\partial x_j} + \frac{\partial u_j'}{\partial x_i} \right) \right\rangle $$

### Its role is directly linked to the evolution of the anisotropy tensor.

### -> Connection to the Machine Learning Target

### The machine learning objective in this work is to model the pressure–strain correlation $\Phi_{ij}$.

### Since $\Phi_{ij}$ governs the evolution of $b_{ij}$, the anisotropy tensor provides physically meaningful input features.

### Understanding the structure of $b_{ij}$ across the channel is therefore essential for developing accurate and stable machine learning closures.

### Summary

#### - The anisotropy tensor quantifies deviations from isotropic turbulence
#### - Strong anisotropy exists near the wall
#### - Pressure-strain drives return to isotropy
#### - Accurate modeling of $\Phi_{ij}$ requires anisotropy-aware features

### This section forms the core physical motivation for the machine learning model.

---

## Analysis 7 : Pressure-Strain Correlation

### The pressure-strain correlation is a central term in the Reynolds stress transport equation and governs the redistribution of turbulent kinetic energy among the velocity components.

### It is defined as:

### $$ \Phi_{ij} = \left\langle \frac{p'}{\rho} \left(\frac{\partial u_i'}{\partial x_j}+\frac{\partial u_j'}{\partial x_i}\right)\right\rangle$$

### Unlike production and dissipation terms, pressure-strain does not create or destroy turbulent kinetic energy, but redistributes it between different components.

### -> Normal Pressure-Strain Components

### The diagonal components $\Phi_{11}$, $\Phi_{22}$, and $\Phi_{33}$ represent redistribution of energy among the three velocity components.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["phi_11"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["phi_11"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["phi_11"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["phi_11"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\Phi_{11}$")
plt.title("Pressure–strain component $\Phi_{11}$")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Shear Pressure-Strain Component

### The off-diagonal component $\Phi_{12}$ redistributes energy associated with turbulent momentum transport.

```python
plt.figure()
plt.plot(df_550["y^+"], df_550["phi_12"], label=r"$Re_\lambda=550$")
plt.plot(df_1000["y^+"], df_1000["phi_12"], label=r"$Re_\lambda=1000$")
plt.plot(df_2000["y^+"], df_2000["phi_12"], label=r"$Re_\lambda=2000$")
plt.plot(df_5200["y^+"], df_5200["phi_12"], label=r"$Re_\lambda=5200$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\Phi_{12}$")
plt.title("Pressure–strain component $\Phi_{12}$")
plt.legend()
plt.grid(True)
plt.show()
```

### -> Energy Redistribution Mechanism

### Pressure-strain correlation redistributes turbulent kinetic energy without changing the total amount of energy:

### $$ \Phi_{11} + \Phi_{22} + \Phi_{33} = 0 $$

### This property ensures conservation of turbulent kinetic energy while enabling transfer between velocity components.

### -> Tensor Symmetry Properties

### The pressure-strain tensor is symmetric:

### $$ \Phi_{ij} = \Phi_{ji} $$

### As a result, only a subset of tensor components are independent.
### For turbulent channel flow, the non-zero independent components are:

### $$ \Phi_{11}, \quad \Phi_{22}, \quad \Phi_{33}, \quad \Phi_{12} $$

### -> Why Only Four Components Are Modeled

### In fully developed channel flow:

#### - Statistical homogeneity in the streamwise and spanwise directions
#### - Symmetry about the channel centerline
#### - Zero mean velocity gradients except $\mathrm{d}U/\mathrm{d}y$

### These constraints eliminate or relate several tensor components, leaving only four independent pressure-strain terms.

### Modeling these four components is sufficient to capture the full pressure-strain dynamics of the flow.

### -> Reynolds Number Dependence

### Comparison across Reynolds numbers shows:

#### - Near-wall behavior exhibits strong Reynolds-number dependence
#### - Outer-layer behavior collapses reasonably well under inner scaling
#### - Higher $Re_\lambda$ increases the range of energetic interactions

### This highlights the importance of scale-aware modeling approaches.

### -> Connection to Machine Learning

### The pressure-strain correlation is the most challenging term to model in Reynolds stress closures due to its non-local and nonlinear nature.

### By learning $\Phi_{ij}$ directly from DNS data using physically informed features, machine learning offers a promising alternative to traditional closure models.

### This section defines the target variables for the learning problem.

### Summary

#### - Pressure-strain redistributes turbulent kinetic energy
#### - It preserves the total turbulent kinetic energy
#### - Only four independent components exist for channel flow
#### - Accurate modeling of $\Phi_{ij}$ is critical for turbulence closure

### This section defines the core prediction target of the machine learning model.

## Analysis 8 : Reynolds Stress Transport Budget: Physical Intuition

### The Reynolds stress transport equation describes the evolution of each Reynolds stress component through competing physical mechanisms:

### $$ \frac{D R_{ij}}{D t} = P_{ij}+\Phi_{ij}-\varepsilon_{ij}+T_{ij}$$

### Each term represents a distinct physical process and dominates in different regions of the turbulent channel flow.

### This section provides **intuition only**, without mathematical derivations.

### -> Production Term $P_{ij}$

### The production term represents the transfer of energy from the mean flow to turbulent fluctuations.

### In channel flow, production arises primarily due to the mean shear:

### $$ P_{12} \propto - \langle u' v' \rangle \frac{\mathrm{d}U}{\mathrm{d}y} $$

### Production is strongest in the buffer layer where both shear and velocity fluctuations are large.

### -> Pressure-Strain Correlation $\Phi_{ij}$

### The pressure-strain term redistributes turbulent kinetic energy between velocity components without creating or destroying energy.

### $$ \Phi_{11} + \Phi_{22} + \Phi_{33} = 0 $$

### It is responsible for:

#### - Return to isotropy
#### - Redistribution of streamwise energy to wall-normal and spanwise components
#### - Stabilization of turbulence structure

### -> Dissipation Term $\varepsilon_{ij}$

### Dissipation represents the irreversible conversion of turbulent kinetic energy into internal energy due to molecular viscosity.

### $$ \varepsilon = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33} $$

### Dissipation is:

#### - Always positive
#### - Strongest near the wall
#### - Smallest in the outer region

### -> Transport Term $T_{ij}$

### Transport represents spatial redistribution of Reynolds stresses due to:

#### - Turbulent diffusion
#### - Pressure diffusion
#### - Viscous diffusion

### Transport does not generate or destroy turbulence but redistributes it across the wall-normal direction.

### -> Dominant Physical Mechanisms by Region

### - **Near-wall region ($y^+ < 30$):**
  #### - Strong dissipation
  #### - Pressure–strain redistributes excess streamwise energy

### - **Buffer layer ($30 < y^+ < 100$):**
  #### - Production dominates
  #### - Maximum energy transfer from mean flow to turbulence

### - **Outer region ($y^+ > 100$):**
  #### - Transport balances pressure–strain
  #### - Reduced dissipation

### -> Why Turbulence Closure Is Difficult

#### - Production depends on mean gradients
#### - Dissipation depends on small-scale dynamics
#### - Transport is non-local
#### - Pressure–strain is nonlinear, non-local, and tensorial

### No single scalar quantity can represent these effects.

### This motivates data-driven tensor-based closure models.

### Summary

#### - Reynolds stress evolution is governed by competing physical mechanisms
#### - Each term dominates in different wall-normal regions
#### - Pressure–strain plays a critical stabilizing role
#### - Accurate closure requires tensor-level modeling

### This section completes the physical foundation for machine learning.

## Analysis 9 : Feature-Target Relationship Exploration (ML Intuition)

### Before training any machine learning model, it is critical to understand how the target quantity - the pressure-strain correlation tensor $\Phi_{ij}$ - relates to the selected input features.

### This section bridges turbulence physics and machine learning intuition by visually inspecting the relationships between:

#### - Reynolds stress anisotropy tensor components $b_{ij}$
#### - Mean velocity gradient $\dfrac{\mathrm{d}U}{\mathrm{d}y}$
#### - Turbulent kinetic energy $k$
#### - Dissipation rate $\varepsilon$

### and the target tensor components $\Phi_{ij}$.

### The goal is to assess:
#### - Linearity vs nonlinearity
#### - Scale dependence
#### - Suitability of different ML model classes

### 🔹Pressure–Strain vs Reynolds Stress Anisotropy

### The pressure–strain correlation is fundamentally responsible for redistributing energy among Reynolds stress components.
### Therefore, a strong relationship with the anisotropy tensor

### $$ b_{ij} = \frac{R_{ij}}{2k} - \frac{1}{3}\delta_{ij} $$
### is expected.

```python
dfs = {
    550: df_550,
    1000: df_1000,
    2000: df_2000,
    5200: df_5200
}

components = [11,22,33,12]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, comp in zip(axes, components):
    for Re, dfi in dfs.items():
        ax.scatter(
            dfi[f"b_{comp}"],
            dfi[f"phi_{comp}"],
            s=10,
            alpha=0.4,
            label=rf"$Re_\lambda={Re}$"
        )
    ax.set_xlabel(rf"$b_{{{comp}}}$")
    ax.set_ylabel(rf"$\Phi_{{{comp}}}$")
    ax.set_title(rf"$\Phi_{{{comp}}}$ vs $b_{{{comp}}}$")
    ax.legend()

plt.tight_layout()
plt.show()
```

### -> **Observations**

#### - Strong nonlinear relationships are observed
#### - Near-wall data clusters differ from outer-layer data
#### - No single linear mapping exists across the entire domain

### This confirms that linear eddy-viscosity or linear tensor mappings are insufficient for modeling $\Phi_{ij}$.

### 🔹Pressure–Strain vs Mean Velocity Gradient

### The mean velocity gradient

### $$ \frac{\mathrm{d}U}{\mathrm{d}y}$$

### drives turbulence production and indirectly influences pressure–strain redistribution.

```python
dfs = {
    550: df_550,
    1000: df_1000,
    2000: df_2000,
    5200: df_5200
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, comp in zip(axes, components):
    for Re, dfi in dfs.items():
        ax.scatter(
            dfi["dU/dy"],
            dfi[f"phi_{comp}"],
            s=10,
            alpha=0.6,
            label=rf"$Re_\lambda={Re}$"
        )
    ax.set_xlabel(r"$\dfrac{\mathrm{d}U}{\mathrm{d}y}$")
    ax.set_ylabel(rf"$\Phi_{{{comp}}}$")
    ax.set_title(rf"$\Phi_{{{comp}}}$ vs $\dfrac{{\mathrm{{d}}U}}{{\mathrm{{d}}y}}$")
    ax.legend()

plt.tight_layout()
plt.show()
```

### -> **Observations**

#### - The relationship is highly scattered and nonlinear
#### - Identical velocity gradients can produce very different $\Phi_{ij}$
#### - Indicates missing state information when using gradients alone

### This motivates the inclusion of turbulence state variables such as $b_{ij}$ and $k$.

### 🔹Pressure-Strain vs Turbulent Kinetic Energy

### The turbulent kinetic energy

### $$ k = \frac{1}{2}\left(R_{11} + R_{22} + R_{33}\right)$$

### sets the energetic scale of turbulence.

```python
dfs = {
    550: df_550,
    1000: df_1000,
    2000: df_2000,
    5200: df_5200
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, comp in zip(axes, components):
    for Re, dfi in dfs.items():
        ax.scatter(
            dfi["k"],
            dfi[f"phi_{comp}"],
            s=10,
            alpha=0.6,
            label=rf"$Re_\lambda={Re}$"
        )
    ax.set_xlabel(r"$k$")
    ax.set_ylabel(rf"$\Phi_{{{comp}}}$")
    ax.set_title(rf"$\Phi_{{{comp}}}$ vs $k$")
    ax.legend()

plt.tight_layout()
plt.show()
```

### -> **Observations**

#### - Larger magnitudes of $\Phi_{ij}$ occur at higher $k$
#### - However, $k$ alone does not determine tensor structure
#### - Directional redistribution information is missing

### Thus, $k$ acts as a **scale variable**, not a closure variable.

### 🔹Pressure–Strain vs Dissipation Rate

### The dissipation rate

### $$\varepsilon = \varepsilon_{11} + \varepsilon_{22} + \varepsilon_{33}$$

### controls small-scale turbulence decay and time scales.

```python
dfs = {
    550: df_550,
    1000: df_1000,
    2000: df_2000,
    5200: df_5200
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, comp in zip(axes, components):
    for Re, dfi in dfs.items():
        ax.scatter(
            dfi["epsilon"],
            dfi[f"phi_{comp}"],
            s=10,
            alpha=0.6,
            label=rf"$Re_\lambda={Re}$"
        )
    ax.set_xlabel(r"$\varepsilon$")
    ax.set_ylabel(rf"$\Phi_{{{comp}}}$")
    ax.set_title(rf"$\Phi_{{{comp}}}$ vs $\varepsilon$")
    ax.legend()

plt.tight_layout()
plt.show()
```

### **Observations**

#### - Tensor-aware ML architectures

### This analysis confirms that pressure–strain modeling is a **data-driven nonlinear regression problem constrained by physics**.

## Analysis 11 : Cross-Reynolds Generalization

### A central requirement for any turbulence closure model is the ability to generalize across different Reynolds numbers.

### For data-driven closures, this translates into the following question:

#### > Can a machine learning model trained at one Reynolds number
#### > accurately predict turbulence behavior at another?

### This section investigates which turbulence quantities exhibit Reynolds-number universality under inner scaling, and which retain explicit Reynolds-number dependence.

### -> Inner-Scaled Variables Across $Re_\lambda$

### We compare the same physical quantities across all available Reynolds numbers:

### $$ Re_\lambda = \{550,\;1000,\;2000,\;5200\} $$

### All quantities are plotted as functions of the inner-scaled wall coordinate $y^+$.

```python
datasets = {
    "550": df_550,
    "1000": df_1000,
    "2000": df_2000,
    "5200": df_5200
}

plt.figure(figsize=(8, 6))

for label, dfi in datasets.items():
    plt.plot(dfi["y^+"], dfi["U"], label=rf"$Re_\lambda = {label}$")

plt.xscale("log")
plt.xlabel(r"$y^+$")
plt.ylabel(r"$U^+$")
plt.title(r"Inner-scaled mean velocity profile")
plt.legend()
plt.grid(True)
plt.show()
```

### -> **Observation**

#### - Excellent collapse in the viscous sublayer and logarithmic region
#### - Confirms classical inner-scaling universality

```python
plt.figure(figsize=(8, 6))

for label, dfi in datasets.items():
    plt.plot(dfi["y^+"], dfi["b_12"], label=rf"$Re_\lambda = {label}$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$b_{12}$")
plt.title(r"Reynolds stress anisotropy across $Re_\lambda$")
plt.legend()
plt.show()
```

### -> **Observation**

#### - Near-wall collapse is strong
#### - Outer-layer deviations increase with $Re_\lambda$
#### - Indicates partial universality

```python
plt.figure(figsize=(8, 6))

for label, dfi in datasets.items():
    plt.plot(dfi["y^+"], dfi["phi_12"], label=rf"$Re_\lambda = {label}$")

plt.xlabel(r"$y^+$")
plt.ylabel(r"$\Phi_{12}$")
plt.title(r"Pressure–strain correlation across $Re_\lambda$")
plt.legend()
plt.show()
```

### -> **Observation**

#### - Strong Reynolds-number dependence
#### - Inner scaling alone is insufficient
#### - Indicates non-universal redistribution physics

### -> Successes and Failures of Inner Scaling

### **Inner scaling succeeds for:**
#### - Mean velocity $U^+$
#### - Near-wall Reynolds stresses
#### - Qualitative trends of anisotropy

### **Inner scaling fails for:**
#### - Pressure–strain correlation $\Phi_{ij}$
#### - Outer-layer anisotropy behavior
#### - High-order statistics

### -> Machine Learning Generalization Requirements

### A successful ML closure must:

#### - Learn universal near-wall physics
#### - Adapt to Reynolds-number-dependent outer-layer behavior
#### - Infer redistribution dynamics not captured by scalar scaling
#### - Maintain physical realizability across all $Re_\lambda$

### -> Universal vs Reynolds-Dependent Quantities

| Quantity | Universality | Comment |
|--------|-------------|---------|
| $U^+$ | High | Classical law of the wall |
| $b_{ij}$ | Partial | Near-wall universal, outer layer dependent |
| $k$ | Partial | Magnitude varies with $Re_\lambda$ |
| $\varepsilon$ | Low | Strong small-scale sensitivity |
| $\Phi_{ij}$ | Low | Highly non-universal redistribution |

### -> Key Takeaway

### Cross-Reynolds generalization is **not automatic** for turbulence closures.

### While inner scaling collapses first-order statistics, pressure–strain redistribution remains strongly Reynolds-number dependent.

### This necessitates:
#### - Rich feature sets
#### - Nonlinear models
#### - Careful cross-Reynolds validation

### These findings directly inform the machine learning strategy developed in the next phase.
